```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
```

```python
nemo = cv2.imread('Testing_Images/leaf2.jpg')
plt.imshow(nemo)
plt.show()
```

```python
nemo = cv2.cvtColor(nemo, cv2.COLOR_BGR2RGB)
plt.imshow(nemo)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib import colors
```

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib import colors
```

```python
r, g, b = cv2.split(nemo)
fig = plt.figure()
axis = fig.add_subplot(1, 1, 1, projection="3d")
```

```python
pixel_colors = nemo.reshape((np.shape(nemo)[0]*np.shape(nemo)[1], 3))
norm = colors.Normalize(vmin=-1.,vmax=1.)
norm.autoscale(pixel_colors)
pixel_colors = norm(pixel_colors).tolist()
```

```python
axis.scatter(r.flatten(), g.flatten(), b.flatten(), facecolors=pixel_colors, marker=".")
axis.set_xlabel("Red")
axis.set_ylabel("Green")
axis.set_zlabel("Blue")
plt.show()
```

```python
hsv_nemo = cv2.cvtColor(nemo, cv2.COLOR_RGB2HSV)
```

```python
h, s, v = cv2.split(hsv_nemo)
fig = plt.figure()
axis = fig.add_subplot(1, 1, 1, projection="3d")

axis.scatter(h.flatten(), s.flatten(), v.flatten(), facecolors=pixel_colors, marker=".")
axis.set_xlabel("Hue")
axis.set_ylabel("Saturation")
axis.set_zlabel("Value")
plt.show()
```

```python
#for leaf2
light_orange = (31,7,149)
dark_orange = (74,216,255)


#for checking 2
#light_orange = (31,21,28) #green
#dark_orange = (117,255,255)#green

#light_orange = (69, 170, 188)
#dark_orange = (69, 240, 29)
```

```python
from matplotlib.colors import hsv_to_rgb
```

```python
lo_square = np.full((10, 10, 3), light_orange, dtype=np.uint8) / 255.0
do_square = np.full((10, 10, 3), dark_orange, dtype=np.uint8) / 255.0
```

```python

plt.subplot(1, 2, 1)
plt.imshow(hsv_to_rgb(do_square))
plt.subplot(1, 2, 2)
plt.imshow(hsv_to_rgb(lo_square))
plt.show()
```

```python
mask = cv2.inRange(hsv_nemo, light_orange, dark_orange)
```

```python
result = cv2.bitwise_and(nemo, nemo, mask=mask)
```

```python

plt.subplot(1, 2, 1)
plt.imshow(mask, cmap="gray")
plt.subplot(1, 2, 2)
plt.imshow(result)
plt.show()
```

```python
#cv2.imwrite('Segmented_Checking_Result.jpg', result)
#cv2.imwrite('Segmented_Checking_Mask.jpg', mask)
cv2.imwrite('Segmented_Result/Segmented_leaf_Mask_FirstMask.jpg', cv2.cvtColor(mask, cv2.COLOR_RGB2BGR)) 
cv2.imwrite('Segmented_Result/Segmented_leaf_Result_FirstMask.jpg', cv2.cvtColor(result, cv2.COLOR_RGB2BGR))
```

```python
#light_white = (0, 0, 200)
#dark_white = (135, 150, 255)
light_white = (14, 33, 74)
dark_white = (115, 255, 255)

#tesult with leaf2.jpg
#light_white = (0, 0, 200)
#dark_white = (145, 200, 255)

#result 3
#light_white = (0, 0, 50) #most deteccting
#dark_white = (145, 150, 255)
```

```python
lw_square = np.full((10, 10, 3), light_white, dtype=np.uint8) / 255.0
dw_square = np.full((10, 10, 3), dark_white, dtype=np.uint8) / 255.0

plt.subplot(1, 2, 1)
plt.imshow(hsv_to_rgb(lw_square))
plt.subplot(1, 2, 2)
plt.imshow(hsv_to_rgb(dw_square))
plt.show()
```

```python
mask_white = cv2.inRange(hsv_nemo, light_white, dark_white)
result_white = cv2.bitwise_and(nemo, nemo, mask=mask_white)

plt.subplot(1, 2, 1)
plt.imshow(mask_white, cmap="gray")
plt.subplot(1, 2, 2)
plt.imshow(result_white)
plt.show()
```

```python
#cv2.imwrite('result6.jpg', result_white)
cv2.imwrite('Segmented_Result/Segmented_leaf_RESULT_White.jpg', cv2.cvtColor(result_white, cv2.COLOR_RGB2BGR))
```

```python
final_mask = mask + mask_white

final_result = cv2.bitwise_and(nemo, nemo, mask=final_mask)
plt.subplot(1, 2, 1)
plt.imshow(final_mask, cmap="gray")
plt.subplot(1, 2, 2)
plt.imshow(final_result)
plt.show()
```

```python
blur = cv2.GaussianBlur(final_result, (7, 7), 0)
plt.imshow(blur)
plt.show()
```
